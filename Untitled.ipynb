```python
image5 = cv2.imread('image4.5.png') 
image6 = cv2.imread('image5.png')
# cv2.bitwise_and is applied over the
# image inputs with applied parameters 
dest_and = cv2.bitwise_and(image5, image6, mask = None)
  
# the window showing output image
# with the Bitwise AND operation
# on the input images
#cv2.imshow('Bitwise And', dest_and)
plt.imshow(dest_and)
```

```python
image5 = cv2.imread('image4.5.png') 
image6 = cv2.imread('image5.png')
dest_or = cv2.bitwise_or(image5, image6, mask = None)
  
# the window showing output image
# with the Bitwise OR operation
# on the input images
#cv2.imshow('Bitwise OR', dest_or)
plt.imshow(dest_or)
```

```python
image = cv2.imread('group.jpg')
  
cv2.imshow('Original Image', image)
cv2.waitKey(0)
  
# Gaussian Blur
Gaussian = cv2.GaussianBlur(image, (25, 25), 0)
cv2.imshow('Gaussian Blurring', Gaussian)
cv2.waitKey(0)
  
# Median Blur
median = cv2.medianBlur(image, 15)
cv2.imshow('Median Blurring', median)
cv2.waitKey(0)
  
  
# Bilateral Blur
bilateral = cv2.bilateralFilter(image, 9, 800, 800)
cv2.imshow('Bilateral Blurring', bilateral)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
